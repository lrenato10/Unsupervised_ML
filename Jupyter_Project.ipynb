# Exercice 2.2 (Interest of K-means)

```python
import  time

nmax_cluster=5 #2 to 5 K
n_metrics=3
data_set = ["xclara.arff", "triangle1.arff"]
n_dataset = len(data_set)

scores = np.zeros((n_dataset, nmax_cluster-1,n_metrics))
for i in range(n_dataset):
    #Import dataset
    path = '/home/lurodrig/Unsupervised_ML/artificial/'

    print('Dataset = ' + str(data_set[i]))
    databrut = arff . loadarff ( open ( path + data_set[i] , 'r') )
    datanp = [ [ x [ 0 ] ,x [ 1 ] ] for x in databrut [ 0 ] ]
    datanp_array=np.array(datanp)
    #Extract Columns 0 and 1 
    f0 = datanp_array[:, 0] # tous les elements de la premiere colonne
    f1 = datanp_array[: ,1] # tous les elements de la deuxieme colonne

    for k in range(2,nmax_cluster+1):
        startTime=time.time()
        print ( "Appel KMeans pour une valeur fixee de k" )
        tps1 = time.time()
        model = cluster.KMeans(n_clusters =k , init = 'k-means++' )
        model.fit( datanp_array )

        tps2 = time.time()
        labels = model.labels_
        iteration = model.n_iter_
        plt.scatter ( f0 , f1 , c = labels , s = 8 )
        plt.title ( " Donnees apres clustering Kmeans " )
        plt.show()
        print (" nb clusters = " ,k , " , nb iter = " , iteration , " runtime = " , round (( tps2 - tps1 ) * 1000 , 2 ) ," ms " )

        scores[i, k-2, 0] = metrics.silhouette_score(datanp_array, labels, metric='euclidean', sample_size=None, random_state=None)
        scores[i, k-2, 1] = metrics.davies_bouldin_score(datanp_array, labels)
        scores[i, k-2, 2] = metrics.calinski_harabasz_score(datanp_array, labels)
        print('Silhouette_score: ' + str(scores[i, k-2, 0]))
        print('davies_bouldin_score: ' + str(scores[i, k-2, 1]))
        print('calinski_harabasz_score: ' + str(scores[i, k-2, 2]))

        executionTime= (time.time()-startTime)
        print('Execution time: ' + str(executionTime) + ' sec')

print('Cluster = [2,3,4,5]')
print('Metric = [silhouette, davies_bouldin, calinski_harabasz]')

for i in range(n_dataset):
    print('Scores dataset Cluster_number x Metric for dataset: ' +str(data_set[i]))
    print(scores[i, :, :])
```
